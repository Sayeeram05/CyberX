```python
# ── Cell 2: IEEE Section III.A Text ───────────────────────────────────────────

IEEE_SECTION_TEXT = """
─────────────────────────────────────────────────────────────────────────────
  READY TO PASTE → IEEE Paper  ·  Section III.A  System Architecture
─────────────────────────────────────────────────────────────────────────────

III. SYSTEM ARCHITECTURE

A. Overview

CyberX is a five-module, full-stack cybersecurity platform built on Django 6.0
(Python 3.12+) that unifies email validation, URL threat detection, phishing
detection, malware analysis, and network intrusion detection within a single
web-accessible dashboard. Fig. 1 depicts the layered system architecture,
organized into five horizontal layers progressing from user-facing inputs to
persistent model and data storage.

Layer 1 (Input) accepts five distinct artifact types: an email address (RFC
5322 string), a URL or hyperlink (string), a phishing-candidate URL (string),
a binary executable or document file (.exe, .pdf, etc.), and raw or live
network traffic in PCAP format. Each input type maps directly to a dedicated
processing module, enabling complete isolation of detection logic and
preventing cross-contamination of feature spaces.

Layer 2 (Interface) comprises the Django web frontend and a set of REST API
endpoints. The dashboard is rendered via server-side HTML/CSS/JavaScript
templates stored under App/Frontend/ and extends a unified base.html layout.
REST views (App/*/views.py) return JSON-serialized responses and apply
@csrf_exempt decorators for programmatic API access. This layer provides a
single unified entry point for both browser-based and programmatic clients.

Layer 3 (Module Apps) contains five isolated Django application modules, each
encapsulating its own models, views, URL routes, and migration history. The
modules operate independently: Email Validation employs a nine-layer
rule-based engine; URL Threat Detection uses a six-step hybrid pipeline;
Phishing Detection applies an 87-feature deep neural network; Malware
Analysis orchestrates a three-engine detection framework; and Network IDS
performs 78-feature bidirectional flow classification.

Layer 4 (Detection Pipelines) is the core inference layer. The Email module
traverses nine verification stages—RFC format, disposable-domain blocklist
(5,100+ entries), WHOIS domain age, SPF, DKIM, DMARC, and MX/DNS—producing a
weighted risk score (0–100). The URL module applies blocklist filtering, domain
whitelisting, structural analysis, reputation heuristics, and a Decision
Tree + Random Forest + Extra Trees majority-vote ensemble (>=95% accuracy).
The Phishing module extracts 87 lexical and structural features, scales via
MinMaxScaler, and forwards them through a PyTorch MLP (87→300→100→1) with
Sigmoid output (~95% accuracy). The Malware module chains MD5/SHA hash lookup,
ten heuristic rules, 41-feature extraction, and an RF (60%) + Gradient
Boosting (40%) weighted ensemble (100% on synthetic benchmark). The Network
IDS module captures PCAP/live packets via Scapy, assembles bidirectional flows
from 5-tuples, computes 78 statistical features, and applies an RF + XGBoost
soft-voting classifier for 7-class intrusion detection (98–100% on CICIDS2017).

Layer 5 (Storage) persists results and model artifacts. Email logs are stored
in SQLite3 (EmailValidationLog, BehavioralFlag). URL, Malware, and NIDS
modules load scikit-learn models from .joblib files. Phishing Detection loads
a PyTorch state dictionary (.pth) alongside a MinMaxScaler. Label encoders
(.json) and feature-name manifests ensure reproducibility.

A key architectural novelty is lazy model loading—artifacts are instantiated
once on first request and cached in memory, minimising inference latency.
Graceful degradation ensures that if a model artifact is unavailable, the
rule-based components continue to operate, maintaining service availability.

─────────────────────────────────────────────────────────────────────────────
  Word count: ~390 words  (IEEE target: 300–400)
─────────────────────────────────────────────────────────────────────────────
"""

print(IEEE_SECTION_TEXT)
```

```python
# ── Cell 5: IEEE-Quality Architecture Diagram (Fig. 1) ────────────────────────

# ─── Color Palette ────────────────────────────────────────────────────────────
C = {
    "bg":              "#F7F9FC",
    "title_bg":        "#1A365D",
    "title_text":      "#FFFFFF",
    "title_sub":       "#90CDF4",

    "email":           "#2B6CB0",
    "email_light":     "#DBEAFE",
    "email_mid":       "#3182CE",

    "url":             "#92400E",
    "url_light":       "#FEF3C7",
    "url_mid":         "#B45309",

    "phishing":        "#6B46C1",
    "phishing_light":  "#EDE9FE",
    "phishing_mid":    "#7C3AED",

    "malware":         "#9B1C1C",
    "malware_light":   "#FEE2E2",
    "malware_mid":     "#DC2626",

    "nids":            "#065F46",
    "nids_light":      "#D1FAE5",
    "nids_mid":        "#059669",

    "django":          "#9A3412",
    "django_light":    "#FFEDD5",

    "storage_bg":      "#EDF2F7",
    "layer_border":    "#A0AEC0",
    "layer_label":     "#2D3748",
    "text_dark":       "#1A202C",
}

MKEYS   = ["email",     "url",              "phishing",        "malware",       "nids"]
MLABELS = ["Email\nValidation", "URL Threat\nDetection", "Phishing\nDetection",
           "Malware\nAnalysis", "Network\nIDS"]
MINPUTS = ["Email Address\n(string)", "URL / Link\n(string)",
           "Phishing URL\n(string)", "Binary File\n(.exe / .pdf)",
           "Network Traffic\n(PCAP / Live)"]
MSUBS   = ["9-layer rule engine", "6-step pipeline",
           "87-feat. neural net", "3-engine analyzer", "78-feat. ensemble"]
MSTORE  = [
    "SQLite3 DB\nEmailValidationLog\nBehavioralFlag",
    "URL Models (.joblib)\nDT + RF + Extra Trees",
    "Phishing (.pth+.joblib)\nPyTorch MLP\nMinMaxScaler",
    "Malware (.joblib)\nRF + GB + Scaler",
    "NIDS (.joblib+.json)\nVotingClassifier\nLabelEncoder",
]

PIPELINES = {
    "email":    ["RFC\nCheck", "Domain\nBlocklist", "WHOIS\nAge",
                 "SPF\nRecord", "DKIM\nSig.", "DMARC\nPolicy", "MX/DNS\nLookup"],
    "url":      ["URL\nNormalize", "Blocklist\nFilter", "Domain\nWhitelist",
                 "Structure\nAnalysis", "Reputation\nScore", "DT+RF+ET\nEnsemble", None],
    "phishing": ["Trusted\nDomain Chk", "Feature\nExtract (87)", "MinMax\nScaling",
                 "PyTorch MLP\n87→300→100→1", "Sigmoid\nThreshold", None, None],
    "malware":  ["Hash\nSignature", "Heuristic\nAnalyzer", "Feature\nExtract (41)",
                 "RF+GB\nEnsemble", "Verdict\nScoring", None, None],
    "nids":     ["PCAP/Live\nCapture", "Scapy\nPkt Parse", "Bidir Flow\nAssembly",
                 "78-Feature\nExtraction", "StandardScaler\nNorm",
                 "RF+XGBoost\nSoft Vote", "7-Class\nPrediction"],
}

# ─── Geometry ─────────────────────────────────────────────────────────────────
FIG_W, FIG_H = 34, 28
COL_W   = 5.6
COL_GAP = 0.9
MARGIN  = 1.05
COL_X   = [MARGIN + i * (COL_W + COL_GAP) for i in range(5)]

STEP_H   = 1.52
STEP_GAP = 0.30

# ─── Drawing Helpers ──────────────────────────────────────────────────────────
def fbox(ax, x, y, w, h, fc, ec=None, lw=1.8, ls="-", pad=0.22,
         alpha=1.0, zorder=3, radius="round"):
    ec = ec or fc
    p = FancyBboxPatch((x, y), w, h, boxstyle=f"{radius},pad={pad}",
                       facecolor=fc, edgecolor=ec, linewidth=lw,
                       linestyle=ls, alpha=alpha, zorder=zorder)
    ax.add_patch(p)
    return p

def tlabel(ax, x, y, text, color, fs, fw="bold", va="center", ha="center",
           style="normal", zorder=5, alpha=1.0):
    ax.text(x, y, text, ha=ha, va=va, color=color, fontsize=fs,
            fontweight=fw, fontstyle=style, zorder=zorder, alpha=alpha,
            multialignment="center")

def arrow_down(ax, x, y_start, y_end, color, lw=1.8, zorder=6):
    ax.annotate("", xy=(x, y_end), xytext=(x, y_start),
                arrowprops=dict(arrowstyle="-|>", color=color,
                                lw=lw, mutation_scale=14), zorder=zorder)

def draw_module_card(ax, x, y, w, h, title, sub, mk):
    fbox(ax, x, y, w, h, fc=C[mk], lw=2.2, pad=0.25, zorder=4)
    tlabel(ax, x + w/2, y + h/2 + h*0.1, title, "white", fs=10.5, zorder=5)
    tlabel(ax, x + w/2, y + h*0.22, sub, "white", fs=8,
           fw="normal", style="italic", zorder=5, alpha=0.88)

def draw_pipeline_step(ax, x, y, w, h, num, text, mk):
    fc_lt = C[mk + "_light"]
    fc_dk = C[mk + "_mid"]
    fbox(ax, x, y, w, h, fc=fc_lt, ec=fc_dk, lw=2.0, pad=0.15, zorder=4)
    badge = plt.Circle((x + 0.45, y + h/2), 0.32, color=fc_dk, zorder=5)
    ax.add_patch(badge)
    tlabel(ax, x + 0.45, y + h/2, str(num), "white", fs=7.5, zorder=6)
    tlabel(ax, x + w/2 + 0.22, y + h/2, text, C["text_dark"], fs=8.2, zorder=5)

def layer_boundary(ax, x, y, w, h, tag, tc):
    fbox(ax, x, y, w, h, fc="none", ec=C["layer_border"],
         lw=1.0, ls="--", pad=0.0, alpha=1.0, zorder=1)
    tlabel(ax, x + 0.35, y + h - 0.28, tag, tc, fs=9, fw="bold",
           style="italic", ha="left", zorder=2)

# ─── Canvas ───────────────────────────────────────────────────────────────────
fig = plt.figure(figsize=(FIG_W, FIG_H), facecolor=C["bg"])
ax  = fig.add_axes([0, 0, 1, 1])
ax.set_xlim(0, FIG_W);  ax.set_ylim(0, FIG_H)
ax.axis("off");          ax.set_facecolor(C["bg"])

# ── TITLE BAND ────────────────────────────────────────────────────────────────
fbox(ax, 0, 25.85, FIG_W, 2.15, fc=C["title_bg"], ec="none",
     lw=0, pad=0, radius="square", zorder=2)
tlabel(ax, FIG_W/2, 27.1,
       "CyberX  —  Full System Architecture",
       C["title_text"], fs=24, zorder=3)
tlabel(ax, FIG_W/2, 26.22,
       "5 Security Modules  ·  Django 6.0  ·  Python 3.12+  ·  "
       "PyTorch  ·  scikit-learn  ·  XGBoost  ·  Scapy",
       C["title_sub"], fs=10.5, fw="normal", zorder=3)

# ── LAYER 1 — USER INPUTS  (y: 23.45 – 25.85) ────────────────────────────────
layer_boundary(ax, 0.4, 23.45, FIG_W - 0.8, 2.25,
               "  LAYER 1  ·  USER INPUTS", C["layer_label"])
for i, mk in enumerate(MKEYS):
    fbox(ax, COL_X[i], 23.65, COL_W, 1.6, fc=C[mk], lw=2.0, pad=0.2, zorder=4)
    tlabel(ax, COL_X[i] + COL_W/2, 23.65 + 0.80,
           MINPUTS[i], "white", fs=9, zorder=5)

for i, mk in enumerate(MKEYS):
    arrow_down(ax, COL_X[i] + COL_W/2, 23.65, 23.30, C[mk])

# ── LAYER 2 — DJANGO FRONTEND  (y: 21.15 – 23.3) ─────────────────────────────
layer_boundary(ax, 0.4, 21.15, FIG_W - 0.8, 2.0,
               "  LAYER 2  ·  DJANGO FRONTEND + REST API  (App/)", C["django"])
fbox(ax, 1.1, 21.35, 14.8, 1.60, fc=C["django"], lw=2.2, pad=0.22, zorder=4)
tlabel(ax, 1.1 + 14.8/2, 21.35 + 0.92,
       "Web Dashboard (Browser)", "white", fs=11, zorder=5)
tlabel(ax, 1.1 + 14.8/2, 21.35 + 0.38,
       "HTML/CSS/JS  ·  base.html templates  ·  App/Frontend/",
       "white", fs=8.5, fw="normal", style="italic", zorder=5, alpha=0.88)

fbox(ax, 17.4, 21.35, 15.85, 1.60, fc=C["django"], lw=2.2, pad=0.22, zorder=4)
tlabel(ax, 17.4 + 15.85/2, 21.35 + 0.92,
       "REST API Endpoints", "white", fs=11, zorder=5)
tlabel(ax, 17.4 + 15.85/2, 21.35 + 0.38,
       "JSON responses  ·  @csrf_exempt  ·  App/*/views.py  ·  App/*/urls.py",
       "white", fs=8.5, fw="normal", style="italic", zorder=5, alpha=0.88)

for i, mk in enumerate(MKEYS):
    arrow_down(ax, COL_X[i] + COL_W/2, 21.35, 21.0, C[mk])

# ── LAYER 3 — MODULE APPS  (y: 19.0 – 21.0) ──────────────────────────────────
layer_boundary(ax, 0.4, 19.0, FIG_W - 0.8, 1.9,
               "  LAYER 3  ·  DJANGO MODULE APPS  (App/)", C["layer_label"])
for i, mk in enumerate(MKEYS):
    draw_module_card(ax, COL_X[i], 19.15, COL_W, 1.55, MLABELS[i], MSUBS[i], mk)

for i, mk in enumerate(MKEYS):
    arrow_down(ax, COL_X[i] + COL_W/2, 19.15, 18.80, C[mk])

# ── LAYER 4 — DETECTION PIPELINES  (y: 4.5 – 18.8) ──────────────────────────
layer_boundary(ax, 0.4, 4.5, FIG_W - 0.8, 14.2,
               "  LAYER 4  ·  DETECTION & ML INFERENCE PIPELINES", C["layer_label"])

PIPE_START = 18.62
for i, mk in enumerate(MKEYS):
    cx    = COL_X[i]
    steps = PIPELINES[mk]
    fc_dk = C[mk + "_mid"]
    fc_lt = C[mk + "_light"]

    # Column header
    tlabel(ax, cx + COL_W/2, PIPE_START + 0.12,
           f"[ {MLABELS[i].replace(chr(10), ' ')} ]",
           fc_dk, fs=9, fw="bold", zorder=5)

    # Column background tint
    col_h = len(steps) * (STEP_H + STEP_GAP) - STEP_GAP + 0.15
    fbox(ax, cx - 0.08, PIPE_START - col_h - 0.05,
         COL_W + 0.16, col_h + 0.08,
         fc=fc_lt, ec=fc_dk, lw=1.0, pad=0.08,
         alpha=0.30, zorder=2)

    y_cur = PIPE_START
    num   = 1
    prev_bot = None
    for step in steps:
        y_top = y_cur - STEP_H
        if step is not None:
            draw_pipeline_step(ax, cx, y_top, COL_W, STEP_H, num, step, mk)
            if prev_bot is not None:
                arrow_down(ax, cx + COL_W/2, prev_bot, y_top + STEP_H, fc_dk, lw=1.6)
            prev_bot = y_top
            num += 1
        y_cur -= (STEP_H + STEP_GAP)

# Arrows → Layer 5
for i, mk in enumerate(MKEYS):
    n_steps  = len(PIPELINES[mk])
    pipe_bot = PIPE_START - n_steps * (STEP_H + STEP_GAP) + STEP_GAP - 0.25
    arrow_down(ax, COL_X[i] + COL_W/2, max(pipe_bot, 4.62), 4.30, C[mk])

# ── LAYER 5 — STORAGE  (y: 0.4 – 4.3) ───────────────────────────────────────
layer_boundary(ax, 0.4, 0.4, FIG_W - 0.8, 3.68,
               "  LAYER 5  ·  DATA & MODEL STORAGE", C["layer_label"])
for i, mk in enumerate(MKEYS):
    fbox(ax, COL_X[i], 0.58, COL_W, 3.20,
         fc=C[mk], lw=2.2, pad=0.22, alpha=0.92, zorder=4)
    tlabel(ax, COL_X[i] + COL_W/2, 0.58 + 1.60,
           MSTORE[i], "white", fs=8.8, zorder=5)

# ── LEGEND ────────────────────────────────────────────────────────────────────
legend_patches = [
    mpatches.Patch(color=C[mk], label=lbl.replace("\n", " "))
    for mk, lbl in zip(MKEYS, MLABELS)
] + [mpatches.Patch(color=C["django"], label="Django Frontend / API")]

ax.legend(handles=legend_patches,
          loc="lower center", bbox_to_anchor=(0.5, -0.006),
          ncol=6, fontsize=10, framealpha=0.95,
          edgecolor="#CBD5E0", handlelength=1.8, handleheight=1.3)

# ── SAVE ──────────────────────────────────────────────────────────────────────
plt.savefig("cyberx_ieee_architecture.png", dpi=150,
            bbox_inches="tight", facecolor=C["bg"])
plt.show()
print("✓ Fig. 1 saved → cyberx_ieee_architecture.png")
```

## Fig. 2 — Per-Module Pipeline Detail Diagram

Horizontal pipeline diagrams for each of the five CyberX modules with:
- Number-badged steps (circle badge top-left of each box)
- Gradient color ramp from darkest (input) → lightest (output) per module
- Sub-labels showing key technical details (feature counts, thresholds, etc.)
- Suitable for inline figure or supplementary material in IEEE paper

```python
# ── Cell 6: Per-Module Pipeline Detail (Fig. 2) ────────────────────────────────

def pipeline_row(ax, title, stages, base_hex, light_hex, note=""):
    """Draw one horizontal pipeline row with gradient-ramp boxes."""
    n = len(stages)

    def blend(c1, c2, t):
        def h2r(h): return tuple(int(h.lstrip("#")[k:k+2], 16)/255 for k in (0,2,4))
        r1,g1,b1 = h2r(c1); r2,g2,b2 = h2r(c2)
        return (r1+(r2-r1)*t, g1+(g2-g1)*t, b1+(b2-b1)*t)

    ax.set_xlim(-0.3, n)
    ax.set_ylim(0, 1.45)
    ax.axis("off")
    ax.set_title(title, fontsize=9.8, fontweight="bold", pad=5,
                 color="#1A202C", loc="left", x=0.01)

    BOX_W, BOX_H, BOX_Y = 0.88, 0.72, 0.36

    for i, item in enumerate(stages):
        if isinstance(item, tuple):
            stage, sub = item
        else:
            stage, sub = item, None

        t  = i / max(n - 1, 1)
        fc = blend(base_hex, light_hex, t * 0.6)
        tc = "white" if t < 0.45 else "#1A202C"

        p = FancyBboxPatch((i, BOX_Y), BOX_W, BOX_H,
                           boxstyle="round,pad=0.04",
                           facecolor=fc, edgecolor="white", linewidth=1.6, zorder=3)
        ax.add_patch(p)

        # Number badge
        badge = plt.Circle((i + 0.13, BOX_Y + BOX_H - 0.13), 0.11,
                            color="white", alpha=0.35, zorder=4)
        ax.add_patch(badge)
        ax.text(i + 0.13, BOX_Y + BOX_H - 0.13, str(i+1),
                ha="center", va="center", fontsize=6.5,
                fontweight="bold", color="white", zorder=5)

        # Main label
        ax.text(i + BOX_W/2, BOX_Y + BOX_H/2 + (0.07 if sub else 0),
                stage, ha="center", va="center",
                color=tc, fontsize=8.0, fontweight="bold",
                multialignment="center", zorder=5)
        if sub:
            ax.text(i + BOX_W/2, BOX_Y + 0.1, sub,
                    ha="center", va="center", color=tc, fontsize=6.5,
                    multialignment="center", zorder=5, alpha=0.85)

        # Arrow →
        if i < n - 1:
            ax.annotate("", xy=(i + BOX_W + 0.04, BOX_Y + BOX_H/2),
                        xytext=(i + BOX_W, BOX_Y + BOX_H/2),
                        arrowprops=dict(arrowstyle="-|>", color="#555555",
                                        lw=1.4, mutation_scale=10), zorder=6)

# ── Pipeline stage data ────────────────────────────────────────────────────────
PIPE_DATA = {
    "email": {
        "title": "① Email Validation Module  (9-Layer Rule-Based Pipeline  ·  Weighted Risk Score 0–100)",
        "base": "#1E4E8C", "light": "#BEE3F8",
        "stages": [
            ("RFC 5322\nRegex",       "Format valid."),
            ("email-\nvalidator",     "IDN + norm."),
            ("Disposable\nBlocklist", "5,100+ doms."),
            ("Heuristic\nEngine",     "14R + 19KW"),
            ("WHOIS\nDomain Age",     "7-day cache"),
            ("SPF\nRecord",           "DNS TXT"),
            ("DKIM\nSig.",            "11 selectors"),
            ("DMARC\nPolicy",         "_dmarc DNS"),
            ("MX / DNS\nCheck",       "RFC 5321"),
            ("Risk\nScore 0–100",     "Weighted sum"),
        ],
    },
    "url": {
        "title": "② URL Threat Detection Module  (6-Step Hybrid Pipeline  +  DT + RF + ET Majority-Vote Ensemble  ·  ≥95% Accuracy)",
        "base": "#7C2D12", "light": "#FDE68A",
        "stages": [
            ("URL\nNormalise",      "Decode + clean"),
            ("Blocklist\n+ IP Chk", "IP regex"),
            ("Domain\nWhitelist",   "100+ trusted"),
            ("Structure\nAnalysis", "20 URL feats."),
            ("Reputation\nScore",   "Heuristics"),
            ("DT + RF + ET\nMajority Vote", "3-model ens."),
            ("Verdict\n+ Risk Score",       "0–100"),
        ],
    },
    "phishing": {
        "title": "③ Phishing Detection Module  (87-Feature PyTorch MLP  ·  87→300→100→1  ·  Sigmoid > 0.5  ·  ~95% Accuracy)",
        "base": "#553C9A", "light": "#E9D8FD",
        "stages": [
            ("Trusted Dom.\nBypass",    "26+ domains"),
            ("Feature\nExtract",        "87 features"),
            ("MinMax\nScaler",          "Fit on train"),
            ("Linear(300)\nBN + ReLU",  "Hidden 1"),
            ("Linear(100)\nBN + ReLU",  "Hidden 2"),
            ("Dropout(0.1)\nLinear(1)", "Regularise"),
            ("Sigmoid\n> 0.5 = Phish",  "~95% acc."),
        ],
    },
    "malware": {
        "title": "④ Malware Analysis Module  (3-Engine: MD5/SHA Signature  +  10 Heuristic Rules  +  RF 60% + GB 40%  ·  41 Features)",
        "base": "#7F1D1D", "light": "#FED7D7",
        "stages": [
            ("File Upload\n.exe/.pdf…",  "Binary input"),
            ("Hash\nSignature",          "MD5 / SHA256"),
            ("Heuristic\n10 Rules",      "PE + entropy"),
            ("Feature\nExtract",         "41 features"),
            ("StandardScaler\nNorm",     "Z-score"),
            ("RF (60%) +\nGB (40%)",     "Weighted ens."),
            ("Composite\nThreat Score",  "0–100"),
            ("Verdict\nClean / Mal.",    "Thresholded"),
        ],
    },
    "nids": {
        "title": "⑤ Network IDS Module  (78-Feature Bidirectional Flow Classification  ·  RF + XGBoost Soft Vote  ·  7 Classes  ·  98–100%)",
        "base": "#064E3B", "light": "#A7F3D0",
        "stages": [
            ("PCAP Upload\nor Live Sniff", "Scapy async"),
            ("Scapy\nParse",               "Pkt fields"),
            ("Bidir Flow\nAssembly",        "5-tuple key"),
            ("78-Feature\nCompute",         "Stats + time"),
            ("StandardScaler\nNorm",        "Z-score"),
            ("RF + XGBoost\nSoft Vote",     "Prob. avg."),
            ("7-Class\nPrediction",         "98–100% acc."),
            ("Async\nPolling",              "Task queue"),
        ],
    },
}

# ── Figure ────────────────────────────────────────────────────────────────────
fig2 = plt.figure(figsize=(28, 16), facecolor="white")
fig2.suptitle(
    "CyberX — Per-Module Pipeline Detail  (Fig. 2)\n"
    "Sequential processing stages for each of the five CyberX security modules",
    fontsize=15, fontweight="bold", y=1.02
)

gs = gridspec.GridSpec(5, 1, figure=fig2, hspace=0.60)
for ri, mk in enumerate(["email", "url", "phishing", "malware", "nids"]):
    ax_r = fig2.add_subplot(gs[ri])
    d    = PIPE_DATA[mk]
    pipeline_row(ax_r, d["title"], d["stages"], d["base"], d["light"])

plt.tight_layout()
plt.savefig("cyberx_ieee_pipeline_detail.png", dpi=150,
            bbox_inches="tight", facecolor="white")
plt.show()
print("✓ Fig. 2 saved → cyberx_ieee_pipeline_detail.png")
```
